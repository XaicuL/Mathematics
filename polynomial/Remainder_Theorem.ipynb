```
# Ex) 1

import sympy as sp

# 1. 기호 변수 선언 (x, a, b)
x, a, b = sp.symbols('x a b')

# 2. 좌변(LHS)과 우변(RHS) 정의
lhs = a*(x-1)**2 + b*(x-1) - 2
rhs = x**2 - 2*x - 1

# 3. 항등식은 '좌변 - 우변 = 0'이 모든 x에 대해 성립해야 함
# 모든 x에 대해 성립하도록 하는 a, b를 구함
equation = lhs - rhs
solution = sp.solve(sp.Poly(equation, x).coeffs(), (a, b))

print(f"a, b = {solution}")

# sp.Poly(equation, x) : x에 대한 다항식 임을 인식
# .coeffs() : 계수들만 도출
# sp.solve(계수, (a, b)) : 위에서 뽑아낸 계수들이 모두 0이 되게 만드는 a와 b의 값 도출
```

```
#항등식
# Ex) 1

import sympy as sp

x,a,b,c = sp.symbols('x,a,b,c,')

lhs1 = x**2 + a*x +  4
rhs1 = (b*x+1)*(x+c)

equ1 = lhs1 - rhs1

re1 = sp.solve(sp.Poly(equ1,x).all_coeffs(), (a,b,c))
print(f'a1,b1,c1 = {re1}')

lhs2 = x ** 2 + 3*x + 5
rhs2 = a*(x+1)**2 + b*(x+1) + c

equ2 = lhs2 - rhs2

re2 = sp.solve(sp.Poly(equ2,x).all_coeffs(), (a,b,c))

print(f'a2,b2,c2 = {re2}')
```

```
import sympy as sp

class PolynomialAnalyzer:
    def __init__(self):
        self.x = sp.Symbol('x')

    def get_poly(self, coeffs):

        return sp.Poly.from_list(coeffs, gens=self.x)

    def verify_factor_theorem(self, poly_obj, a):

        remainder = poly_obj.subs(self.x, a)
        is_factor = (remainder == 0)
        return is_factor, remainder

    def get_quotient_and_remainder(self, p1, p2):

        q, r = sp.div(p1, p2)
        return q, r


analyzer = PolynomialAnalyzer()

model_poly = analyzer.get_poly([1, -6, 11, -6])

is_f, rem = analyzer.verify_factor_theorem(model_poly, 2)
print(f"인수 여부: {is_f}, r: {rem}")
```

# 조립제법

```
import sympy as sp

def synthetic_division(coeffs, a):
    """
    다항식 coeffs를 (x - a)로 나눌 때의 조립제법
    coeffs: [x^n 계수, x^(n-1) 계수, ..., 상수항]
    a: (x - a)에서의 a 값
    """
    n = len(coeffs)
    # 결과를 담을 리스트 (몫의 계수들과 마지막 나머지)
    result = [0] * n
    # 중간 곱셈 과정을 담을 리스트 (시각화용)
    steps = [0] * n

    # 1. 첫 번째 계수는 그대로 내려온다
    result[0] = coeffs[0]

    # 2. 조립제법 루프 (내려오고 -> 곱하고 -> 더하고)
    for i in range(1, n):
        steps[i] = result[i-1] * a  # 앞의 결과에 a를 곱함
        result[i] = coeffs[i] + steps[i] # 원래 계수와 더함

    quotient = result[:-1] # 마지막을 제외한 나머지가 몫의 계수
    remainder = result[-1] # 마지막 값이 나머지

    return quotient, remainder, steps

# --- 실행 예시: P(x) = x^3 - 10x + 3 을 (x - 3)으로 나누기 ---
coeffs = [1, 0, -10, 3]
a_val = 3

q, r, s = synthetic_division(coeffs, a_val)

print(f"원래 계수: {coeffs}")
print(f"곱셈 과정: {s}")
print(f"최종 계수: {q} | 나머지: {r}")
```

## 제수가 1차식인 경우

```
# Ex) 5

import sympy as sp

def synthetic_division_advanced(coeffs, divisor_a, divisor_b):
    """
    다항식을 (divisor_a * x + divisor_b)로 나눌 때의 조립제법
    예: (2x - 1) -> divisor_a=2, divisor_b=-1
    """
    # 조립제법에 사용할 값 alpha = -b/a
    alpha = -divisor_b / divisor_a

    n = len(coeffs)
    result = [0] * n

    # 1. 조립제법 기본 루프 수행 (x - alpha로 나눈 효과)
    result[0] = coeffs[0]
    for i in range(1, n):
        step = result[i-1] * alpha
        result[i] = coeffs[i] + step

    raw_quotient = result[:-1]
    remainder = result[-1]

    # 2. 핵심: 몫을 divisor_a로 나누어 진짜 몫을 구함
    final_quotient = [q / divisor_a for q in raw_quotient]

    return final_quotient, remainder

# --- 테스트: (4x^3 - 4x^2 - 3x + 3) 을 (2x - 1)로 나누기 ---
coeffs = [4, -4, -3, 3]
# 2x - 1 이니까 a=2, b=-1
q, r = synthetic_division_advanced(coeffs, 2, -1)

print(f"몫의 계수: {q}") # 결과: [2.0, -1.0, -2.0] -> 2x^2 - x - 2
print(f"나머지: {r}")         # 결과: 2.0
```
