# 인수분해

```python
import sympy as sp

x = sp.Symbol('x')
# 예: 4x^3 - 4x^2 - 3x + 3
poly = 4*x**3 - 4*x**2 - 3*x + 3

# 인수분해 실행
factored_poly = sp.factor(poly)
print(f"인수분해 결과: {factored_poly}")
```

## 공통부분 인수분해

```python
import sympy as sp

x = sp.symbols('x')

poly = x**4 -3*x**2 -4

factored_poly = sp.factor(poly)

print(f"인수분해 결과: {factored_poly}")
```

```python
import sympy as sp

x, y = sp.symbols('x,y')

poly = (x+y)*(x+y+6)+9

factored_poly = sp.factor(poly)

print(f"인수분해 결과: {factored_poly}")
```

```python
# Ex) 1

import sympy as sp

x = sp.symbols('x')

poly = x**4 -10*x**2 + 9

factored_poly = sp.factor(poly)

print(f"인수분해 결과: {factored_poly}")
```

```python
# Ex) 2

import sympy as sp

x = sp.symbols('x')

poly = (x**2 + x)**2 + 2*(x**2 + x) - 8

factored_poly = sp.factor(poly)

print(f"인수분해 결과: {factored_poly}")
```
